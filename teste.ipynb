```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# String de conexão com o banco de dados Access
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=dados/base.accdb;'
    r'charset=utf-8;'
)
```

```python
# Conectar ao banco de dados
conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

# Consulta SQL para buscar empresas com base nos critérios de consulta
sql = "SELECT * FROM adm;"

df = pd.read_sql_query(sql, conn)
```

```python
# Plot a pie chart
df
```

```python
df['pergunta_1'] = df['pergunta_1'].apply(lambda x : 1 if x == 'bom' else 2)
```

```python
# Count the number of audits for each sector
sector_counts = df['setor_auditado'].value_counts()
```

```python
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sector_counts, labels=sector_counts.index, autopct='%1,1f', startangle=140)
plt.title('Distribution of Audits by Sector')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
